### 1. What is name of car that has the highest horsepower?  

```python
#your code here
df = pd.read_csv("Cars.csv")
df.columns = df.columns.str.strip()
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
max_hp = df['horsepower'].max()
highest_hp_cars = df[df['horsepower'] == max_hp][['name', 'horsepower']]
print(highest_hp_cars)
```

### 2. How many cars have mpg  35?

```python
#your code here
high_mpg_count = df[df['mpg'] >= 35].shape[0]
print("Number of cars with mpg  35:", high_mpg_count)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
filtered_df = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]
most_common_origin = filtered_df['origin'].mode()[0]
print("Most common origin for cars with horsepower > 100 and weight < 3000 is ", most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
japan_cars = df[df['origin'].str.strip().str.lower() == 'japan']
mean_acceleration = round(japan_cars['acceleration'].mean(), 2)
print("Mean acceleration of cars from Japan:", mean_acceleration)

        
```

### 5. Which year had the highest average mpg?

```python
#your code here
avg_mpg_per_year = df.groupby('model_year')['mpg'].mean()
year_highest_avg_mpg = avg_mpg_per_year.idxmax()
highest_avg_mpg = avg_mpg_per_year.max()
print(f"The year with the highest average mpg: {int(year_highest_avg_mpg)} (Average mpg: {highest_avg_mpg:.2f})")
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
mpg_median = df['mpg'].median()
above_median_mpg = df[df['mpg'] > mpg_median].copy()
above_median_mpg['hp_to_weight'] = above_median_mpg['horsepower'] / above_median_mpg['weight']
max_ratio = above_median_mpg['hp_to_weight'].max()
best_cars = above_median_mpg[above_median_mpg['hp_to_weight'] == max_ratio]
print("Car(s) with best horsepower-to-weight ratio (among cars with above-median mpg):")
print(best_cars[['horsepower', 'weight', 'mpg', 'hp_to_weight']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()
plt.figure(figsize=(12, 7))
sns.set_style("whitegrid")
sns.lineplot(
    data=avg_mpg,
    x='model_year',
    y='mpg',
    hue='origin',
    palette='viridis',  
    marker='o',
    markersize=8,
    linewidth=2.5
)
plt.title('Evolution of Average MPG by Origin (1970-1982)', fontsize=18, pad=20)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average MPG', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Origin', title_fontsize=13, fontsize=12, loc='best')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
df = pd.read_csv('Cars.csv')
df.columns = df.columns.str.strip().str.lower()
if 'name' not in df.columns:
    raise KeyError("Expected column 'name' not found in the CSV.")
df['normalized_name'] = df['name'].str.strip().str.lower()
multi_year_models = df.groupby('normalized_name').filter(
    lambda x: x['model_year'].nunique() > 1
)
stats = (
    multi_year_models
    .groupby('normalized_name')
    .agg(
        count=('mpg', 'count'),
        mpg_std=('mpg', 'std'),
        avg_mpg=('mpg', 'mean'),
        years_produced=('model_year', pd.Series.nunique)
    )
)
consistent_models = stats[(stats['mpg_std'] < 1.0) & (stats['count'] >= 2)]
sorted_models = consistent_models.sort_values(
    by=['count', 'avg_mpg'], ascending=[False, False]
)
print(sorted_models.reset_index())
```
